Sin tener en cuenta las restricciones se podrían repetir operaciones y signos, y por lo tanto nos encontraríamos ante un problema de variación con repetición, es decir, queremos tomar k elementos dentro de un conjunto de n elementos y contar las posibles elecciones, teniendo en cuenta que, en este caso, puede
elegirse más de una vez un mismo elemento y que el orden de elección importa, el número de posibilidades se calcularía como: 

                                                 VRn,k = n^k
                                                
Según el número de cifras, n_c = 9, y el valor de k_c iría desde 2 a 9. Por otro lado, el número de operadores es n_o = 4, y tendríamos que elegir k_o = k_c - 1. 

Por lo tanto, el número de posibilidades para las cifras sería: 

                                                VRCn_c,k_c = n_c ^ k_c

donde n_c es número de elementos del conjunto posible de cifras {1 , 2 , 3 , 4 , 5 , 6 , 7 , 8 , 9 } , por lo tanto n_c = 9. Por otro lado, k_c es el número de elementos que escogemos del conjunto, pudiendo haber diferentes posibilidades, concretamente k_c = { 2 , 3 , 4 , 5 , 6 , 7 , 8 , 9 } (mínimo tendríamos que elegir dos cifras). 

Por otro lado, el número de posibilidades para los operadores lo denotamos como: 

                                                VROn_o,k_o = n_o ^ k_o

Concretamente, n_o siempre es 4, ya que el conjunto de operadores es {+ , - , / , * }, mientras que k_o depende del número de cifras elegidas (k_c), dado que siempre vamos a necesitar un signo menos que dicho número, por lo que k_o = k_c - 1. 

En el siguiente código implementamos el cálculo de todas las soluciones si no hubiera restricciones en el problema. Teniendo en cuenta todo lo que hemos comentado anteriormente, este número final sería: 

                                                      (VRCn_c,k_c) ^ (VROn_o,k_o)

Este cálculo es muy complejo según vamos aumentando el valor de las cifras que se pueden elegir, por lo que no está impreso todas las opciones, pero sólo habría que seguir el razonamiento. 

```
# Número de posibilidades de cifras
conjunto_cifras = {1,2,3,4,5,6,7,8,9}
n_c = len(conjunto_cifras)
k_c = {2,3,4,5,6,7,8,9}

# Número de posibilidades de signos
conjunto_signos = {"+","-","/","*"}
n_s = len(conjunto_signos)

print( f"Número total de cifras n_c = {n_c}")
print( f"Número total de signos n_s = {n_s}")

for k in k_c: 
  VRC = (n_c)**(k)
  k_s = k - 1
  VRS = (n_s)**(k_s)
  VR_total = VRC**VRS
  print( "---------")
  print( f"\n Número de cifras usadas k_c = {k}")
  print( f"\n Número de signos usados k_s = {k_s}")
  print( f"\n Número de posibilidades de combinaciones de cifras VRC = {VRC}")
  print( f"\n Número de posibilidades de combinaciones de signos VRS = {VRS}")
  print( f"\n Número de posibilidades total con los parámetros VRT = {VR_total}")
  print( "---------")
```

El inconveniente del algortimo por fuerza bruta es que busca todas las posibles soluciones, es decir, expresiones que dan lugar al número entero que queremos (o desde el otro enfoque, todas las posibles combinaciones entre cifras y signos para sacar el mínimo y máximo valor obtenido). Esto hace que la complejidad sea muy alta como hemos explicado en el punto anterior, además de que la eficiencia temporal es baja.

Por lo tanto, deberíamos mejorar este algoritmo. Para ello, una opción es realizar **ramificación y poda**. Es decir, descartar aquellas ramas que no puedan ofrecer la solución que buscamos (el número entero a obtener). Así, no se terminará la exploración de esa rama, sino que se descartará definitivamente, y con ello disminuirá la complejidad, aumentando la eficiencia del algoritmo. 

Por ejemplo, un criterio para descartar un nodo y sus consiguientes nodos hijos sería que en la permutación formada por los signos y cifras seleccionadas, el número sea difernte al que se necesita. De esta forma, lo descartaremos y no evaluaremos las expresiones. 

También podemos implementar un **algoritmo genético**. 

En este caso, partimos de una población inicial (soluciones candidatas) cuyo tamaño es un parámetro que podemos definir, por ejemplo 50 o 100. Así, cada individuo que la conforma es una posible solución candidata. Cabe destacar que esta expresión cumple las condiciones del problema (usar 4 operadores sin repetir, 5 cifras sin repetir y ambos alternados). 

Comenzaremos con una primera aproximación del algoritmo genético para encontrar la expresión generadora de un número entero. Como sabemos, la función fitness nos permite evaluar a individuo, es decir, obtenemos un resultado sobre la habilidad de dicho individuo para competir con otros de la polación. Por ello, teniendo en cuenta nuestro objetivo, la función fitness será evaluar la expresión (individuo) para comprobar si se obtiene o no el número entero que se busca. Esto lleva a que el fitness score sólo podrá ser 0 (False) si da otro número entero diferente al buscado, o 1 (True) si gracias a esa expresión obtenemos el valor deseado. 

Este proceso de evaluación con la función fitness se llevará a cabo para todos los individuos de la población. 
Si se encuentra un individuo cuyo fitness score sea 1, se detendrá la ejecución del algoritmo y habremos llegado a la solución final. 
Sin embargo, si tras evaluar la población entera no hemos encontrado ningún individuo con un fitness score de 1, procederemos ha realizar cruces y mutaciones para generar una nueva población.

Suponemos que nos encontramos en el segundo caso. Comenzaremos realizando un cruce de dos individuos. Para ello, los seleccionamos al azar de la población inicial (dado que todos tienen fitness score de 0), y los denominamos "padre1" y "padre2". A continuación los cruzamos en múltiples puntos, es decir, el individuo "hijo" tendrá los 4 operadores de sus padres, mientras que las 5 cifras serán aleatorias de sus padres. 

Después, mutamos este individuo, lo cuál significa que volvemos a reordenar de forma aleatoria sus caracteres. 

Realizamos este proceso de forma iterativa hasta formar la nueva población del mismo tamaño que la original, y la evaluamos tal y como hemos definido para encontrar al individuo buscado. 

Este proceso tendrá lugar hasta llegar a una solución tal y como veremos a continuación en el código implementado. Las funciones principales relacionadas con la base teórica de un algoritmo genético son las siguientes.

```
import random

# Generamos los individuos de la población (expresiones)
def crear_individuo(posibles_cifras: list,operadores: list):
  cifras = random.sample(posibles_cifras, 5)
  expresion = "".join([cifra+operacion for cifra, operacion in zip(cifras, operadores)])
  return expresion+cifras[-1]


# Creamos la población de individuos (tamaño como parámetro de entrada)
def crear_poblacion(posibles_cifras, operadores, n_individuos):
  poblacion = set()
  while len(poblacion) < n_individuos:
    # Generamos un individuo
    individuo = crear_individuo(posibles_cifras,operadores)
    # Comprobamos que genera un número entero
    resultado = eval(individuo)
    if check_int(resultado) :
      poblacion.add(individuo)
  return poblacion

# Función fitness -> el resultado de la expresión es el número entero que buscamos 
def funcion_fitness(resultado, n_entero):
  if int(resultado) == int(n_entero):
    return True    # Sería un 1
  else:
    return False   # Sería un 0


# Crossover de dos individuos para generar un descendiente
def cruzar(individuo1, individuo2):
  # Siempre son iguales
  operadores =  ["+","-","/","*"]
  # Cifras del primer individuo
  cifras1 = (individuo1.translate({ord('+'): " ", ord('-'): " ",ord('*'): " ", ord('/'): " "})).split(" ")
  # Cifras del segundo individuo
  cifras2 = (individuo2.translate({ord('+'): " ", ord('-'): " ",ord('*'): " ", ord('/'): " "})).split(" ")
  todas_cifras = list(set(cifras1+cifras2))
  # Seleccionamos las 5 cifras de forma aleatoria que va a tener el "hijo"
  cifras_cruce = random.sample(todas_cifras, 5)
  # Individuo resultante del crossover
  expresion = "".join([cifra+operacion for cifra, operacion in zip(cifras_cruce, operadores)])
  return expresion+cifras_cruce[-1]


# Aleatorizamos las cifras y operadores para generar una mutación en diversos puntos
def mutacion(individuo):
  # Siempre son iguales
  operadores =  ["+","-","/","*"]
  cifras = (individuo.translate({ord('+'): " ", ord('-'): " ",ord('*'): " ", ord('/'): " "})).split(" ")
  # Aleatorizamos las listas
  random.shuffle(cifras)
  random.shuffle(operadores)
  # Creamos el individuo mutado
  individuo_mutado = ("".join([cifra+operacion for cifra, operacion in zip(cifras, operadores)]))+cifras[-1]
  # Comprobamos que es diferente
  if individuo_mutado != individuo:
    return individuo_mutado
  else:
    mutacion(individuo)


# Seleccionamos dos individuos al azar para ser los "padres"
def seleccionar_padres(index1,n_individuos):
  index2 = indice_aleatorio(n_individuos)
  # Comprobamos que no hemos elegido el mismo
  if indices_diferentes:
    return index2
  else:
    indice_aleatorio(n_individuos)		


# Generamos una nueva población a partir de dos padres
# Para ello se hará un crossover (cruce) y una mutación
def crear_nueva_poblacion(n_individuos,poblacion):
  nueva_poblacion = set()
  # Nueva población del mismo tamaño que la original 
  while len(nueva_poblacion) < n_individuos:
    # Seleccionamos dos padres al azar
    index1 = indice_aleatorio(n_individuos)
    index2 = seleccionar_padres(index1,n_individuos)
    padre1 = list(poblacion)[index1]
    padre2 = list(poblacion)[index2]
    # Cruzamos esos dos individuos
    crossover_individuo = cruzar(padre1,padre2)
    # Mutamos el individuo descendiente
    mutacion_individuo = mutacion(crossover_individuo)
    nueva_poblacion.add(mutacion_individuo)
  return nueva_poblacion


# Proceso principal básico -> evaluamos cada uno de los individuos generados 
def seleccion(poblacion,n_entero):
  solucion_encontrada = False
  # Evaluamos cada individuo
  for individuo in list(poblacion):
    try:
      # Resultado (número) de evaluar la expresión
      resultado = eval(individuo)
    except:
      continue
    # Comprobamos si es el número entero que buscamos
    if funcion_fitness(resultado, n_entero):
      solucion_encontrada = True
      return individuo,solucion_encontrada

# Comprobar que el número es entero
def check_int(resultado):
  if (resultado % 1 == 0):
    return True
  else:
    return False

# Comprobar que se eligen índices diferentes
def indices_diferentes(index1,index2):
  if index1 != index2:
    return True
  else:
    return False

# Generamos un valor entero aleatorio como índice
def indice_aleatorio(n_individuos):
  return random.randint(0,n_individuos-1)
```

Finalmente, creamos una función para implementar el comportamiento deseado de nuestro algoritmo. 

```
# Función principal para encontrar una expresión que genere número entero
def encontrar_n_entero(posibles_cifras, operadores, n_entero, n_individuos):
  # Creamos una población inicial
  poblacion = crear_poblacion(posibles_cifras,operadores,n_individuos)
  # Evaluamos los individuos para encontrar una solución
  individuo, solucion_encontrada = seleccion(poblacion,n_entero)

  # Si no hemos encontrado ningun individuo solución en esta población generamos otra
  while not solucion_encontrada:

    # Creamos la nueva población -> crossover y mutación
    nueva_poblacion = crear_nueva_poblacion(n_individuos,poblacion)
    # Evaluamos los individuos para encontrar una solución
    individuo, solucion_encontrada = seleccion(nueva_poblacion,n_entero)

  return individuo
```

Para comprobar el funcionamiento ponemos el ejemplo de que con una población de 50 individuos, donde los signos y cifras posibles son los determinados al inicio del problema, queremos encontrar la expresión que genere el número entero 4. Como vemos, una expresión sería "6+8-2/1*5".

Así,hemos encontrado un algoritmo que en menos de 0,2 segundos encuentra la solución deseada.

```
posibles_cifras = ["1","2","3","4","5","6","7","8","9"]
operadores = ["+","-","/","*"]
n_entero = "4"
n_individuos = 50

expresion_generadora = encontrar_n_entero(posibles_cifras, operadores, n_entero, n_individuos)
print(f"La expresión generadora del número entero es {expresion_generadora}")
```

Por otro lado, vamos a implementar una variación de dicho algoritmo genético para encontrar el mínimo y máximo valor que se puede obtener con las expresiones generadas. Además, comprobaremos si se crean todos los números enteros intermedios o no. 

Para ello, debemos establecer un parámetro nuevo al que denominaremos número de iteraciones. Aunque el funcionamiento es tal y como hemos explicado para el otro caso, necesitamos poner un número máximo de iteraciones el cuál hace referencia a cuántas poblaciones tendremos en cuenta para buscar dichos valores. 

Además, ahora la función fitness será únicamente la evaluación de la expresión (individuo), por lo que el fitness score hará referencia al valor entero que genera. 

```
from collections import defaultdict
import numpy as np

# Obtenemos el fitness score de cada individuo
def funcion_fitness(individuo): 
  return eval(individuo)

# Evaluamos los individuos para obtener todos los resultados 
def evaluar_individuos(poblacion,resultados_y_expresiones):
  for individuo in list(poblacion):
    # Fitness score = resultado de evaluar la expresión
    resultado = funcion_fitness(individuo)
    # Guardamos los resultados en un diccionario
    resultados_y_expresiones[resultado].append(individuo)
  return resultados_y_expresiones

# Función principal
def encontrar_min_max_entero(posibles_cifras, operadores, n_individuos, n_iteraciones):
  # Estructura para guardar los resultados -> {n_entero1 = [exp1,exp2...],n_entero2 = [exp3,exp4...]}
  resultados_y_expresiones =  defaultdict(list)
  # Creamos la población inicial del tamaño indicado
  poblacion = crear_poblacion(posibles_cifras,operadores,n_individuos)
  # Evaluamos todos los individuos y guardamos los resultados
  resultados_y_expresiones = evaluar_individuos(poblacion,resultados_y_expresiones)

  # Creamos tantas poblaciones nuevas como definamos
  for i in range(n_iteraciones):
    # Nueva población como resultado de crossover y mutación
    nueva_poblacion = crear_nueva_poblacion(n_individuos,poblacion)
    # Evaluamos todos los individuos y guardamos los resultados
    resultados_y_expresiones = evaluar_individuos(poblacion,resultados_y_expresiones)
  
  return resultados_y_expresiones


def min_entero_encontrado(resultados):
  return min(resultados.keys())

def max_entero_encontrado(resultados):
  return max(resultados.keys())

def check_todos_enteros(resultados,min_entero,max_entero):
  todos_enteros = np.arange(min_entero,max_entero,1)
  enteros_generados = list(resultados.keys())
  enteros_no_generados = [x for x in todos_enteros if x not in set(enteros_generados)]
  return len(enteros_generados), len(enteros_no_generados)

```

```
posibles_cifras = ["1","2","3","4","5","6","7","8","9"]
operadores = ["+","-","/","*"]
n_individuos = 50
n_iteraciones = 10

resultados = encontrar_min_max_entero(posibles_cifras, operadores, n_individuos, n_iteraciones)
n_min = min_entero_encontrado(resultados)
n_max = max_entero_encontrado(resultados)
n_enteros_generados_intermedios, n_enteros_faltan = check_todos_enteros(resultados,n_min,n_max)

print(f"El mínimo valor entero encontrado entre {n_individuos*n_iteraciones} es {n_min} y el valor máximo encontrado es {n_max}.")
print(f"Se han generado {n_enteros_generados_intermedios} enteros entre el máximo y el mínimo, por lo que faltarían {n_enteros_faltan}.")
```

Tal y como vemos, el número entero mínimo y máximo depende del número de individuos que consideremos en la población, al igual que del número de generaciones que se crean. 

De forma general tras diferentes pruebas, podríamos concluir que no se encuentran todas los números enteros intermedios entre el máximo y el mínimo de un experimento como este por ejemplo de 100 generaciones descendientes de una población de 50 individuos.
